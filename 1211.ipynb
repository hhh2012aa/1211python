```python
# Standardization(標準化), Normalization 回家自己做
df2['funded_amnt_stand'] = (df2['funded_amnt'] - df2['funded_amnt'].mean())/ (df2['funded_amnt'].std()) 
```

```python
#normalize
scale = StandardScaler()
#reshape 將矩陣做成幾行幾列, -1就是自動計算的意思
df2_list_2D = np.array(df2['funded_amnt']).reshape(-1,1)
df2['funded_amnt_norm'] = pd.DataFrame(scale.fit_transform(df2_list_2D))
```
